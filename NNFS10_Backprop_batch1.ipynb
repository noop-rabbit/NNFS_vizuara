```python
import numpy as np
```

```python
dvalues = np.array([
    [1., 1., 1.],
    [2., 2., 2.],
    [3., 3., 3.],
]
)

inputs = np.array([
    [1., 2., 5.2, 1.3],
    [2.1, 3., 6.7, 9.2],
    [6.5, 3.8, 2, 1.4],
]
)
dweights = np.dot(inputs.T, dvalues)
print(dweights)
```

```python
dvalues = np.array([
    [1., 1., 1.],
    [2., 2., 2.],
    [3., 3., 3.],
]
)

biases = np.array([[1., 2., 3.]])

dbiases = np.sum(dvalues, axis=0, keepdims=True)
print(dbiases, dbiases.shape)
```

```python
dvalues = np.array([
    [1., 1., 1.],
    [2., 2., 2.],
    [3., 3., 3.],
]
)
weights = np.array([
    [1., 2.5, 3.2, 1.3],
    [2.1, 3.6, 1.7, 2.2],
    [3.5, 1.8, 2.1, 5.4],
]
).T

dinputs = np.dot(dvalues, weights.T)
print(dinputs)
```
