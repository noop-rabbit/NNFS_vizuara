```python
import numpy as np
```

```python
d_values = np.array([
    [0.1, 0.5, 0.3, 0.1],
    [0.2, 0.1, 0.4, 0.3],
    [0.3, 0.2, 0.1, 0.4],
]
)
```

```python
y_true = np.array([1, 2, 3])
```

```python
samples = len(d_values)
labels = len(d_values[0])
print(f"Samples: {samples}, labels: {labels}")
```

```python
y_true = np.eye(labels)[y_true]
print(y_true)
```

```python
d_inputs = -y_true /  d_values
print(d_inputs)
```

```python
d_inputs_norm = (d_inputs - d_inputs.min()) / (d_inputs.max() - d_inputs.min())
print(d_inputs_norm)
```
