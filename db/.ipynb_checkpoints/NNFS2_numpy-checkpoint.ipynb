```python
import numpy as np
print(np.__version__)
```

```python
inputs = [1,2,3,4]
wts = [1.3, 0.5, -0.3, 0.89]
b = 2
outputs = np.dot(wts, inputs) + b
print(f"Output:{outputs}")
outputs_r = np.dot(inputs, wts) + b
print(f"Output_r:{outputs_r}")
```

```python
inputs = [1, -2, 3.3]
wts = [[0.2, 1, 2.1],
       [-2, 2.5, 6],
       [8, 1, 3.7]]
b = [2 , 5, -1]
layer_op = np.dot(wts,inputs) + b #correct, refer image above
print(f"layer correct output:{layer_op}")
layer_op_w = np.dot(inputs, wts) + b #wrong, refer image above
print(f"layer wrong output:{layer_op_w}")
```
