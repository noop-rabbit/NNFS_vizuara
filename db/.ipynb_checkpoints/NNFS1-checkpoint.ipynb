```python
inputs = [1,2,3,4]
wt = [0.3, 0.7, -0.2, 0.1]
b = 2

outputs = (inputs[0] * wt[0] 
           + inputs[1] * wt[1] 
           + inputs[2] * wt[2]
           + inputs[3] * wt[3]
           + b) 
print(f"NN output:{outputs}")
```

```python
inputs = [1,2,3,4]
wt = [[0.3, 0.7, -0.2, 0.1],
      [0.8, -0.23, 0.7, 0.14],
      [0.5, -0.7, -0.2, 0.71]]
b = [2, -5, 0.3]

wt0 = wt[0]
wt1 = wt[1]
wt2 = wt[2]

outputs = [
    inputs[0] * wt0[0] 
    + inputs[1] * wt0[1] 
    + inputs[2] * wt0[2]
    + inputs[3] * wt0[3]
    + b[0],
    inputs[0] * wt1[0] 
    + inputs[1] * wt1[1] 
    + inputs[2] * wt1[2]
    + inputs[3] * wt1[3]
    + b[1],
    inputs[0] * wt2[0] 
    + inputs[1] * wt2[1] 
    + inputs[2] * wt2[2]
    + inputs[3] * wt2[3]
    + b[2],
]
    
print(f"NN output:{outputs}")
print(type(outputs))
```

```python
inputs = [1,2,3,4]
wts = [[0.3, 0.7, -0.2, 0.1],
      [0.8, -0.23, 0.7, 0.14],
      [0.5, -0.7, -0.2, 0.71]]
b = [2, -5, 0.3]

layer_outputs = []

for n_wt, n_b in zip(wts, b):
    neuron_op = 0

    for n_ip, wt in zip(inputs, n_wt):
        print(f"Neuron input:{n_ip} , Weight:{wt}")
        neuron_op += n_ip * wt
    print(f"Neuron bias:{n_b}")
    neuron_op += n_b
    layer_outputs.append(neuron_op)
print(f"Final output:{layer_outputs}")
        
```
