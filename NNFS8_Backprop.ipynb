```python
weights = np.array([1.0, 3.0, 5.0])
bias = 1
inputs = np.array([-3.0, 1.0, 2.0])
target = 0
lr = 0.001     #learning_rate
iterations = 10000
loss_last = 0

def leaky_relu(x):
    return np.where(x > 0, x, 0)

def rel_grad(x):
    return np.where(x > 0, 1, 0)

for iteration in range(iterations):
    # Forward pass
    linear_op = np.dot(weights, inputs) + bias
    op = relu(linear_op)
    loss = (op - target) ** 2
    cv = loss_last - loss
    #print(cv)
    if(abs(cv) > 1e-8):
        #Backward pass

        dL_dop = 2 * (op - target)
        dop_dlin = rel_grad(linear_op)
        dlin_dwt = inputs
        dlin_dbias = 1.0
    
        dL_dlin = dL_dop * dop_dlin
        dL_dwt = dL_dlin * dlin_dwt
        DL_dbias = dL_dlin * dlin_dbias
    
        #update weights and bias
        weights -= lr *  dL_dwt
        bias -= lr * DL_dbias
        loss_last = loss
        print(f"iteration: {iteration + 1}, Loss: {loss}" )
    else:
        break
    
print(f"Final weights: {weights}")
print(f"Final bias: {bias}")
    
```
